```python
import pandas as pd
import json

# --- Konfigurasi ---
excel_file_name = "pendidikan_dasar.xlsx"
output_geojson_name = "pendidikan_dasar.geojson"

try:
    # 1. Baca data dari Excel
    df = pd.read_excel(excel_file_name)

    # 2. Bersihkan data
    df.dropna(how='all', inplace=True)
    df['KOORDX'] = pd.to_numeric(df['KOORDX'], errors='coerce')
    df['KOORDY'] = pd.to_numeric(df['KOORDY'], errors='coerce')
    df.dropna(subset=['KOORDX', 'KOORDY'], inplace=True)

    # Ganti NaN dengan None → supaya jadi null di JSON
    df = df.astype(object).where(pd.notna(df), None)

    # 3. Buat daftar features GeoJSON
    features = []
    for _, row in df.iterrows():
        properties = row.drop(['KOORDX', 'KOORDY']).to_dict()

        # Konversi Timestamp ke string
        for key, value in properties.items():
            if isinstance(value, pd.Timestamp):
                properties[key] = str(value)

        longitude = row['KOORDX']
        latitude = row['KOORDY']

        geometry = {
            "type": "Point",
            "coordinates": [longitude, latitude]
        }

        feature = {
            "type": "Feature",
            "geometry": geometry,
            "properties": properties
        }
        features.append(feature)

    # 4. Buat struktur GeoJSON
    geojson_data = {
        "type": "FeatureCollection",
        "features": features
    }

    # 5. Simpan ke file
    with open(output_geojson_name, 'w', encoding='utf-8') as f:
        json.dump(geojson_data, f, indent=2, ensure_ascii=False)

    print(f"✅ File '{output_geojson_name}' berhasil dibuat!")

except FileNotFoundError:
    print(f"❌ Error: File '{excel_file_name}' tidak ditemukan.")
except Exception as e:
    print(f"❌ Terjadi error: {e}")
```

```python
from google.colab import files
import pandas as pd
import json
import os

# Upload beberapa file Excel sekaligus
uploaded = files.upload()

# Buat folder output
os.makedirs("output", exist_ok=True)

for excel_file_name in uploaded.keys():
    try:
        base_name = os.path.splitext(os.path.basename(excel_file_name))[0]
        output_geojson_name = f"output/{base_name}.geojson"

        # 1. Baca data
        df = pd.read_excel(excel_file_name)

        # 2. Bersihkan data
        df.dropna(how='all', inplace=True)
        df['KOORDX'] = pd.to_numeric(df['KOORDX'], errors='coerce')
        df['KOORDY'] = pd.to_numeric(df['KOORDY'], errors='coerce')
        df.dropna(subset=['KOORDX', 'KOORDY'], inplace=True)

        # Ganti NaN dengan None → supaya jadi null di JSON
        df = df.astype(object).where(pd.notna(df), None)

        # 3. Buat daftar features
        features = []
        for _, row in df.iterrows():
            properties = row.drop(['KOORDX', 'KOORDY']).to_dict()
            for key, value in properties.items():
                if isinstance(value, pd.Timestamp):
                    properties[key] = str(value)

            geometry = {
                "type": "Point",
                "coordinates": [row['KOORDX'], row['KOORDY']]
            }

            feature = {
                "type": "Feature",
                "geometry": geometry,
                "properties": properties
            }
            features.append(feature)

        geojson_data = {
            "type": "FeatureCollection",
            "features": features
        }

        # 4. Simpan ke file
        with open(output_geojson_name, 'w', encoding='utf-8') as f:
            json.dump(geojson_data, f, indent=2, ensure_ascii=False)

        print(f"✅ Berhasil konversi: {excel_file_name} → {output_geojson_name}")

    except Exception as e:
        print(f"❌ Error pada file {excel_file_name}: {e}")
```
